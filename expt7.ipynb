```python
data = {
    "Name": ["Anant Singh", "Shabd Kumar", "Sumati Kumar", "Raj Mehta", "Priya Sharma", "Aditi Verma", 
             "Vikram Joshi", "Ravi Singh", "Neha Bansal", "Sneha Gupta"],
    "Experience": [2, 5, 1, 4, 3, 6, 7, 1, 8, 2],
    "Salary": [50000, 70000, 60000, 65000, 55000, 80000, 95000, 45000, 90000, 52000]
}
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df["Name"], df["Salary"], color='skyblue')
plt.title("Salary of ours")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["Experience"], df["Salary"], marker='o', linestyle='-', color='orange')
plt.title("Salary Based on Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df["Salary"], labels=df["Name"], autopct='%1.1f%%', startangle=140)
plt.title("Salary pie chart anant company\n")
plt.axis('equal')
plt.show()
```
